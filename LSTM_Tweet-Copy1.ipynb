```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
# vocab size
voc_size=5000
```

```python
#one hot representation
messages=x.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
#messages.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# data preprocessing removing unnecessory words
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()    # for stmeming
corpus = []     # creating a list of corpus 
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['text'][i]) # replace or substituting every hting with a space apart from a-z and A-Z char
    review = re.sub('http[\w:/\.]+',' ', messages['text'][i]) # remove urls
    
    review = review.lower()     #lower all chars
    review = review.split()     # split for removing stopwords
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]   # if words are not in stopwords then stemming
    review = ' '.join(review)   # join the splits
    corpus.append(review)     # append to list corpus
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]  # one hot for corpus word and vocab #words of corpus cpnvert in index based on vocab word's index
onehot_repr
```

```python
# every sentance has different length but we have to supply fixed length sentances to the model so we do pad sequenceing (appending 0 to make all statements equal length)
# embading representation
sent_length=20  # all sentanses length should be 20 words because we are dealing with title only which is short in nature for long text we can expend it.
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) # pre means add 0 before the stances 
print(embedded_docs)    # it has all x values
```

```python
embedded_docs[0]
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40

model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
#model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np             # convert all data into array so that we can perform further measures
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
x_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, test_size=0.30,random_state=42)   # one more parameter---> , random_state=42
```

```python
y_test.size
```

```python
#training
model.fit(x_train,y_train,validation_split=0.20,batch_size=32,epochs=4)   #validation_split=0.20  or  validation_data=(x_test,y_test)
```

```python
#Performance Metrics And Accuracy
y_pred=model.predict_classes(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python

report=classification_report(y_test,y_pred)
print(report)
#confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
